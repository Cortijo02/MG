```python
import cv2
import matplotlib.pyplot as plt
```

```python
#pics_n = [19, 22, 80, 65, 70, 48, 71, 62, 87, 97]
pics_n = [19,      80, 65, 70,     71,     87]
fig, axs = plt.subplots(figsize=(10, 11), ncols=6, nrows=len(pics_n))
fig.tight_layout()
plt.subplots_adjust(wspace=0, hspace=0.3)

for y in range(len(pics_n)):
      hr = cv2.imread('HR/' + str(pics_n[y]) + '.png')
      lr = cv2.imread('LR/' + str(pics_n[y]) + '.png')
      sr = cv2.imread('SR/' + str(pics_n[y]) + '.png')
      real = cv2.imread('Real-SRGAN/' + str(pics_n[y]) + '.png')
      real = cv2.imread('SwinIR/' + str(pics_n[y]) + '.png')
      real = cv2.imread('ResShift/' + str(pics_n[y]) + '.png')

      axs[y, 0].imshow(hr, cmap='gray')
      axs[y, 0].axis("off")
      if y == 0: axs[y, 0].set_title('HR')

      axs[y, 1].imshow(lr, cmap='gray')
      axs[y, 1].axis("off")
      if y == 0: axs[y, 1].set_title('LR')

      axs[y, 2].imshow(sr, cmap='gray')
      axs[y, 2].axis("off")
      if y == 0: axs[y, 2].set_title('SR')

      axs[y, 3].imshow(real, cmap='gray')
      axs[y, 3].axis("off")
      if y == 0: axs[y, 3].set_title('Real ESRGAN')

      axs[y, 4].imshow(real, cmap='gray')
      axs[y, 4].axis("off")
      if y == 0: axs[y, 4].set_title('SwinIR')

      axs[y, 5].imshow(real, cmap='gray')
      axs[y, 5].axis("off")
      if y == 0: axs[y, 5].set_title('ResShift')
```
